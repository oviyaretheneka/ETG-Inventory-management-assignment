```python
import json
import time
import random
import string
from datetime import datetime
```

### USER DEFINED FUNCTIONS

```python
# file handling 


# record.json open(read mode)

def f_open():
    f=open("record.json","r")
    r=f.read()
    f.close()
    rec=json.loads(r)
    return(rec)


#record.json updation(write mode)

def f_update(rec):
    f=open("record.json","w")
    dum=json.dumps(rec)
    f.write(dum)
    f.close()
```

```python
# generation of random transcation id using random and string modules

def trans_id():
        N=5
        t_id = ''.join(random.choices(string.digits + string.digits, k=N))
        return(t_id)

```

```python
#file handling 

#sales.json open(read mode)

def s_open():
    fd=open("sales.json","r")
    s=fd.read()
    fd.close()
    s_rec=json.loads(s)
    return(s_rec)


#sales.json updation(write mode)

def s_update(s_rec):
    fd=open("sales.json","w")
    fd.write(json.dumps(s_rec))
    fd.close()
```

```python
#file handling


#expiry.json open(read mode)

def ex_open():
    fe=open("expiry.json","r")
    tp=fe.read()
    fe.close()
    tp_expired=json.loads(tp)
    return(tp_expired)

#expiry.json updation(write mode)

def ex_update(tp_expired):
    fe=open("expiry.json","w")
    dum=json.dumps(tp_expired)
    fe.write(dum)
    fe.close()
```

```python
#user define function to avoid repetition of same product name during the new entry of items

def name_check(rec):
    n_name=input("Enter the product name : ")
    for i in rec:
        if(rec[i]["p_name"].lower()==n_name.lower()):
            print("Name already existing")
            print("try again")
            name_check(rec)
    else:
        return(n_name.lower())
```

```python
# user defined function to avoid repetition of same product id during the new entry of items


def id_check(rec):
    n_id=input("Enter the  product id : ")
    if n_id in rec.keys():
            print("Entered product id is clashing with the existing one, enter the new id : ")
            print("try again")
            id_check(rec)
    else:
            return(n_id)
        
```

```python
#user defined function to make a report and to alert the admin about the products


def check_expiryreport(rec):
    tp_expired=ex_open()              #expiry.json file open
    current_time = datetime.now()
    c_month=current_time.month
    rec=f_open()
    expired={}
    a_expired={}
    
    
    
    print("***************************   Inventory Product Report    *****************************")
    print()
    
    # manipulating data of expiry products
    
#      we consider a product is expired after 6 months of the pakage or manufactured date """
    
    
    for i in rec:
        
        if abs(int(rec[i]["pkd"][3:5])-c_month)>6 and abs(int(rec[i]["pkd"][6:])-current_time.year)==0:
            expired[i]=rec[i]["p_name"]
        
        elif(abs(int(rec[i]["pkd"][3:5])-c_month)==6 and abs(int(rec[i]["pkd"][6:])-current_time.year)==0):
            a_expired[i]=rec[i]["p_name"]
            

            
    # generation of report  for expired products and products that are about to be expired\
    
    
    if expired!={}:
        
        print("************ Expired products *************")
        
        for i in expired:
            
            print()
            print("Product",":",rec[i]["p_name"])
            print()
    
    else:
        
        print("""    At present there is no expired products at the inventory   """)
    
    if a_expired != {}:
        
        print("************ Products to be expired within a month  *************")
        
        for i in a_expired:
            
            print()
            print("Product",":",rec[i]["p_name"],"is due for {} days to expiry \n".format(abs(current_time.day-int(rec[i]["pkd"][0:2]))))
    else:
        print("""    At present there is no  products about to be expired at the inventory   """)
            
    
        
    tp_expired["p_expired"]=expired
    tp_expired["pa_expired"]=a_expired
    
    ex_update(tp_expired)                #expiry.json file updation of changes
    
    
    
                
            
    
    
```

```python
#user defined function to alert the admin about the products that are about to be out of stock

def check_quantity():
    rec=f_open()
    alert={}
    for i in rec:
        if rec[i]["quantity_in_stock"]<=5:
            alert[i]=rec[i]["quantity_in_stock"]
   

    if alert != {}:
        
        print("****       ALERT!!!!    ****")
        print()
        print("    product quantity in crisis   ")
        print()
        
        for i in alert:
            
            print("Product",rec[i]["p_name"],":","only has {} items left".format(rec[i]["quantity_in_stock"]))
            print()
        
    
            
            
```

### ADMIN SPACE

```python
#user defined function for admin activities such as 
                         
"""                          1.Updation of inentory details
                             2.Deletion of the whole item
                             3.Creation of a whole new item in the inventory
                             4.Generation of expiry product item                            """


        

def admin():
    print("""Choose your action
                                    1.Update Inventory
                                    2.Deletion of Product                  
                                    3.New Entry        
                                    4.Expiry Report
                                                                 """)
    
    
    
    action=int(input())
    
    
    # updation of inventory
    
    if action==1:
        
        try:
            
                rec=f_open()
                p_id=input("Enter the product id which is needed to be updated : ")
                if p_id in rec.keys():
                        print(""" choose your action 
                                            1.product_name
                                            2.unit price of the item
                                            3.pakage date
                                            4.quantity avaliable in stock
                                                                                     """)
                        choice=int(input())

                        if choice==1:
                                u_name=input("Enter the new product name : ")
                                rec[p_id]["p_name"]=u_name
                        elif(choice==2):
                                u_price=eval(input("Enter the unit price of {} : ".format((rec[p_id])["p_name"])))
                                rec[p_id]["unit_price"]=u_price
                        elif(choice==3):
                                u_pkd=input("Enter the pakage date in the format date/month/year of {} : ".format(rec[p_id]["p_name"]))
                                rec[p_id]["pkd"]=u_pkd

                        elif(choice==4):
                                u_quantity=int(input("Enter the qunatity to be updated : "))
                                rec[p_id]["quantity_in_stock"]=u_quantity
                                rec[p_id]["inventory_value"]=u_quantity*rec[p_id]["unit_price"]
                        else:
                                print("Entry is mismatched with the inventory")

                        print("Your updation is successful !")

                        f_update(rec)
                else:
                    print("Invaild product id")
                    print("Try again")
        except:
            print("Try again error occured !")


     #deletion of item
    
    if action==2:
        
                try:

                    rec=f_open()
                    p_id=input("Enter the product id to delete item from the inventory : ")
                    if p_id in rec.keys():
                        del rec[p_id]
                        print(rec)
                        print("Deletion is successfull")
                        f_update(rec)
                    else:
                        print("Try again product id is unavailabe")
                except:
                    print("Mismatched entry")
                    print("Try Again")

     # insertion of new entry
    
    if action==3:
        
                try:
                    rec=f_open()
                    n_name=name_check(rec)
                    n_id=id_check(rec)
                    n_price=eval(input("Enter the unit price : "))
                    n_quantity=int(input("Enter the qunatity to be added : "))
                    n_pkd_date=input("Enter the package date : ")


                    rec[n_id]={"p_name":n_name,"unit_price":n_price,"quantity_in_stock":n_quantity,"pkd":n_pkd_date,"inventory_value":n_price*n_quantity}
                    f_update(rec)
                    print("New entry is successfully added")

                except:
                    print("Something Went Wrong")
                    print("Try Again")
     
    
    #generation of expiry report
    
    if action==4:
                rec=f_open()
                check_expiryreport(rec)
                
    
```

```python
#user info such as p_id is considered to be an input of product id from the barcode scanner that is been scanned

#but here it is instead typed manually


def cont_check():
    try:
        print(""" Do u want to continue [y/n] where y-->yes
                                                    n-->no       """)
        option=input()
        if option.lower()=="y":
            return(True)
        elif(option.lower()=="n"):
            return(False)
        else:
            print("Invalid input")
        cont_check()
        
    except:
        print("Invalid input")

        
def user():
    
    rec=f_open()         #record.json file open
    
    check_quantity()      #quantity out of stock alert
    
    print()
    
    check_expiryreport(rec)   #expiry of product report
    
    print()
    
    
    u_name=input("Enter the customer name : ")
    u_ph_no=int(input("Enter the user mobile number : "))
    

    lis_pid=[]
    lis_quantity=[]
    
    tp_expired=ex_open()
    
    condition=True
    while condition:
            t_id=trans_id()
            p_id=input("Enter the id")
            
            if p_id in rec.keys():
            
                    quantity=int(input("Enter the total quantity of {} : ".format(rec[p_id]["p_name"])))

                    if rec[p_id]["quantity_in_stock"]<quantity:
                        print("Avaliable quantity is",rec[p_id]["quantity_in_stock"])
                        print("Do you like to purchase the avaliable amount --> [y/n] : ")
                        c=input()
                        if c.lower()=="y":
                            quantity=rec[p_id]["quantity_in_stock"]
                        else:
                            break

                    lis_pid.append(p_id)
                    lis_quantity.append(quantity)


                    if p_id in tp_expired["p_expired"]:
                        print("Product  :",rec[p_id]["p_name"],"is expired already")
                        print("Product is removed from the puchase list")
                        lis_pid.remove(p_id)
                        lis_quantity.remove(quantity)

                    condition=cont_check()

            else:
                
                print("Invalid product id")
                print("Try Again")

                
     # purchase bill generation
    
    print("**********************************************************************************************************")
    print()
    print("   **** **** **** ****  **** **** ****  Hello",u_name,"**** **** **** **** **** **** **** **** **** ")
    print()
    print("                                       Here is your bill                           ")
    print()
    
    
    
    
    lis_bill=[]



     # manipulating the changes made to inventory after the purchase 
        
    for  i in range(len(lis_pid)):
            p_ammount=rec[lis_pid[i]]["unit_price"]*lis_quantity[i]
            
            print(rec[lis_pid[i]]["p_name"],":","  ",p_ammount)
            print()
            
            lis_bill.append(p_ammount)

            if rec[lis_pid[i]]["quantity_in_stock"]-quantity<0:
                rec[lis_pid[i]]["quantity_in_stock"]=0
                
            else:
                rec[lis_pid[i]]["quantity_in_stock"]=rec[lis_pid[i]]["quantity_in_stock"]-quantity

            rec[lis_pid[i]]["inventory_value"]=rec[lis_pid[i]]["inventory_value"]-p_ammount

    print("***************************************************************************************************************")
    print()
    print("                          Total         =",sum(lis_bill))
    print()
    print("***************************************************************************************************************")
    
    
    f_update(rec)





#sales.json --> dumping  all the sales made to customer in a nested dictionary

    s_rec=s_open()


    t_id=trans_id()
    lis_amount=[lis_quantity[x]*rec[lis_pid[x]]["unit_price"] for x in range(len(lis_pid))]
    sales_item={}
    s_namelis=[rec[x]["p_name"] for x in lis_pid ]
    i_names=["p_name","quntity","amount"]
    i_value=[s_namelis,lis_quantity,lis_amount]
    sales_item=dict(zip(i_names,i_value))
    sales_item["t_amount"]=sum(lis_amount)
    s_rec[t_id]=sales_item

    s_update(s_rec)
```
